```python
# LIBRRARY BOOK MANAGEMENT SYSTEM
#Add/remove/search books stored in a CSV file using file handling; functions for each operations;try-expect for file not found errors.
import csv

FILE = "books.csv"


# Add Book
def add_book():
    title = input("Enter book name: ")

    with open(FILE, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([title])

    print("Book Added")


# Remove Book
def remove_book():
    try:
        title = input("Enter book to remove: ")

        with open(FILE, "r") as file:
            books = list(csv.reader(file))

        with open(FILE, "w", newline="") as file:
            writer = csv.writer(file)

            for book in books:
                if book[0] != title:
                    writer.writerow(book)

        print("Book Removed")

    except FileNotFoundError:
        print("File not found")


# Search Book
def search_book():
    try:
        title = input("Enter book to search: ")

        with open(FILE, "r") as file:
            reader = csv.reader(file)

            found = False

            for book in reader:
                if book[0] == title:
                    found = True

            if found:
                print("Book Found")
            else:
                print("Book Not Found")

    except FileNotFoundError:
        print("File not found")


# Menu
while True:
    print("\n1.Add Book")
    print("2.Remove Book")
    print("3.Search Book")
    print("4.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_book()

    elif choice == "2":
        remove_book()

    elif choice == "3":
        search_book()

    elif choice == "4":
        break

    else:
        print("Invalid Choice")
```

```python
#EXPENSE TRACKER
#log daily expenses to a .txt file, read and summarize by category using functions;handles invalid input via exceptional hanlding


def add_expense():
    try:
        category = input("Enter category: ")
        amount = float(input("Enter amount: "))

        file = open("expense.txt", "a")
        file.write(category + " " + str(amount) + "\n")
        file.close()

        print("Expense Added")

    except ValueError:
        print("Invalid amount")


def show_summary():
    total = {}

    try:
        file = open("expense.txt", "r")

        for line in file:
            category, amount = line.split()
            amount = float(amount)

            if category in total:
                total[category] += amount
            else:
                total[category] = amount

        file.close()

        print("\nSummary")
        for i in total:
            print(i, "=", total[i])

    except FileNotFoundError:
        print("File not found")


add_expense()
show_summary()
```

```python
#STUDENT FILEMANAGER SYSTEM
#Read student data from CSV file,computer grades using functions,update results back to CSV files
import csv
import os

FILE_STUDENTS = "students.csv"
FILE_GRADED_STUDENTS = "graded_students.csv"

def initialize_students_csv():
    """Creates a dummy students.csv if it doesn't exist."""
    if not os.path.exists(FILE_STUDENTS):
        with open(FILE_STUDENTS, "w", newline="") as outfile:
            writer = csv.writer(outfile)
            writer.writerow(["ID", "Name", "Marks"])
            writer.writerow(["101", "Alice", "85"])
            writer.writerow(["102", "Bob", "92"])
            writer.writerow(["103", "Charlie", "68"])
        print(f"Created a dummy '{FILE_STUDENTS}'.")

def calculate_grade(marks):
    """Calculates the grade based on marks."""
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    else:
        return "F"

def add_student():
    """Adds a new student record to students.csv."""
    student_id = input("Enter student ID: ")
    name = input("Enter student name: ")
    marks = input("Enter marks: ")

    with open(FILE_STUDENTS, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([student_id, name, marks])
    print(f"Student {name} added to {FILE_STUDENTS}.")

def remove_student():
    """Removes a student record from students.csv by ID."""
    student_id_to_remove = input("Enter ID of student to remove: ")
    students = []
    try:
        with open(FILE_STUDENTS, "r", newline="") as infile:
            reader = csv.reader(infile)
            header = next(reader) # Read header
            students.append(header)
            for row in reader:
                if row[0] != student_id_to_remove:
                    students.append(row)

        with open(FILE_STUDENTS, "w", newline="") as outfile:
            writer = csv.writer(outfile)
            writer.writerows(students)
        print(f"Student with ID {student_id_to_remove} removed from {FILE_STUDENTS}.")
    except FileNotFoundError:
        print(f"Error: '{FILE_STUDENTS}' not found.")

def process_grades():
    """Reads student data, calculates grades, and writes to graded_students.csv."""
    student_data = []
    try:
        with open(FILE_STUDENTS, "r", newline="") as infile:
            reader = csv.reader(infile)
            header = next(reader) # Skip header
            for row in reader:
                student_data.append(row)

        with open(FILE_GRADED_STUDENTS, "w", newline="") as outfile:
            writer = csv.writer(outfile)
            writer.writerow(["ID", "Name", "Marks", "Grade"])

            for student_id, name, marks_str in student_data:
                try:
                    marks = float(marks_str)
                    grade = calculate_grade(marks)
                    writer.writerow([student_id, name, marks_str, grade])
                except ValueError:
                    print(f"Warning: Invalid marks for student {name} (ID: {student_id}). Skipping grade calculation for this student.")
                    writer.writerow([student_id, name, marks_str, "N/A"])
        print(f"Grades processed successfully. Results saved to '{FILE_GRADED_STUDENTS}'.")

    except FileNotFoundError:
        print(f"Error: '{FILE_STUDENTS}' not found. Please ensure the file exists.")
    except Exception as e:
        print(f"An unexpected error occurred during grade processing: {e}")

def view_file_content(filename):
    """Displays the content of a given CSV file."""
    try:
        with open(filename, "r", newline="") as file:
            reader = csv.reader(file)
            print(f"\n--- Content of {filename} ---")
            for row in reader:
                print(', '.join(row))
            print("--------------------------")
    except FileNotFoundError:
        print(f"Error: '{filename}' not found.")

# Main program loop
initialize_students_csv()

while True:
    print("\n--- Student File Manager System ---")
    print("1. Add New Student")
    print("2. Remove Student")
    print("3. Process Grades (and generate graded_students.csv)")
    print("4. View Students Data")
    print("5. View Graded Students Data")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        add_student()
    elif choice == '2':
        remove_student()
    elif choice == '3':
        process_grades()
    elif choice == '4':
        view_file_content(FILE_STUDENTS)
    elif choice == '5':
        view_file_content(FILE_GRADED_STUDENTS)
    elif choice == '6':
        print("Exiting Student File Manager. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
#PASSWORD GENERATOR AND VALIDATOR
#Generates random strong password using random and string modules; validates user entered passwords with custom exception if rule not met


import random
import string


class PasswordError(Exception):
    pass


def generate_password(length=12):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

def validate_password(password):
    if len(password) < 8:
        raise PasswordError("Password must be at least 8 characters long.")

    if not any(char.isupper() for char in password):
        raise PasswordError("Password must contain at least one uppercase letter.")

    if not any(char.islower() for char in password):
        raise PasswordError("Password must contain at least one lowercase letter.")

    if not any(char.isdigit() for char in password):
        raise PasswordError("Password must contain at least one digit.")

    if not any(char in string.punctuation for char in password):
        raise PasswordError("Password must contain at least one special character.")

    return True


generated_password = generate_password()
print("Generated Password:", generated_password)


try:
    user_password = input("Enter a password to validate: ")
    if validate_password(user_password):
        print("Password is valid and strong.")
except PasswordError as e:
    print("Validation Error:", e)


```

```python
#Employee Payroll System
class Employee:
    def __init__(self, name, employee_id, basic_salary):
        self.name = name
        self.employee_id = employee_id
        self.basic_salary = basic_salary
    def calculate_salary(self):
        allowance=self.basic_salary*0.20
        deduction=self.basic_salary*0.10
        final_salary=self.basic_salary+allowance-deduction
        return allowance,deduction,final_salary

employees={}
n=int(input("Enter the number of emplyees: "))
for i in range(n):
  print("Employee",i+1)
  name=input("Enter the name: ")
  employee_id=input("Enter the employee id: ")

  salary = 0.0
  salary_input_valid = False
  while not salary_input_valid:
    try:
      salary=float(input("Basic salary: "))
      if salary<=0:
        raise ValueError("Salary cannot be zero or negative.")
      salary_input_valid = True
    except ValueError as e:
      print(f"Invalid salary: {e}. Please enter a positive number.")

  # Create Employee object and add to dictionary
  employee=Employee(name,employee_id,salary)
  employees[employee_id]=employee


print("EMPLOYEE SALARY REPORT")
for emp_id,emp in employees.items(): # Corrected iteration variable name for clarity
  allowance,deduction,final_salary=emp.calculate_salary()
  print("Employee ID:",emp_id)
  print("Name:",emp.name)
  print("Basic Salary:",emp.basic_salary)
  print("Allowance:",allowance)
  print("Deduction:",deduction)
  print("Final Salary:",final_salary)

```

```python
#Banking system
customers={}
account_numbers=[]
class BankAccount:
  def __init__(self,acc_no,name,balance):
    self.acc_no=acc_no
    self.name=name
    self.balance=balance

  def deposite(self,amount):
    self.balance+=amount
    self.save_transaction(f"Deposited {amount}")
    print("Amount deposited successfully")
  def withdraw(self,amount):
    try:
      if amount>self.balance:
        raise Exception("Insufficient balance")
      self.balance-=amount
      self.save_transaction(f"Withdrew {amount}")
      print("Amount withdrawn successfully")
    except Exception as e:
      print(e)
  def check_balance(self):
    print("Current balance:",self.balance)
  def save_transaction(self,transaction):
    with open(f"{self.acc_no}.txt","a") as file:
      file.write(transaction+"\n")

n=int(input("Enter the number of customers: "))
for i in range(n):
  print("Customer",i+1)
  name=input("Enter the name: ")
  acc_no=input("Enter the account number: ")
  balance=float(input("Enter the balance: "))
  customers[acc_no]=BankAccount(acc_no,name,balance)
  account_numbers.append(acc_no)
while True:
  print("\n1.Deposite")
  print("2.Withdraw")
  print("3.Check Balance")
  print("4.Exit")
  choice_str=input("Enter your choice: ")

  try:
    choice=int(choice_str)
  except ValueError:
    print("Invalid input. Please enter a number from 1 to 4.")
    continue

  if choice==4:
    break

  acc=input("Enter the account number: ")
  if acc not in customers:
    print("Invalid account number")
    continue
  customer=customers[acc]
  if choice ==1:
    amount=float(input("Enter the amount to deposite: "))
    customer.deposite(amount)
  elif choice==2:
    amount=float(input("Enter the amount to withdraw: "))
    customer.withdraw(amount)
  elif choice==3:
    customer.check_balance()
  else:
    print("Invalid choice")

  print("Stored Account Numbers:")
  print(account_numbers)
```

```python
#Hospital management system
class Patient:
    def __init__(self, patient_id, name, doctor):
        self.patient_id = patient_id
        self.name = name
        self.doctor = doctor

    def display(self):
        print("\nPatient ID:", self.patient_id)
        print("Name:", self.name)
        print("Doctor:", self.doctor)


patients = {}

def add_patient():
    try:
        pid = input("Enter Patient ID: ")

        if pid in patients:
            raise Exception("Patient ID already exists")

        name = input("Enter Patient Name: ")
        doctor = input("Enter Doctor Name: ")

        p = Patient(pid, name, doctor)
        patients[pid] = p

        with open("patients.txt", "a") as file:
            file.write(f"{pid},{name},{doctor}\n")

        print("Patient record added successfully")

    except Exception as e:
        print("Error:", e)


def show_patients():
    if not patients:
        print("No records found")
    else:
        for p in patients.values():
            p.display()


while True:
    print("\n1.Add Patient")
    print("2.Show Patients")
    print("3.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_patient()

    elif choice == "2":
        show_patients()

    elif choice == "3":
        break

    else:
        print("Invalid choice")
```

```python
#Student result processing system
students = []

def calculate_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 75:
        return "A"
    elif avg >= 60:
        return "B"
    elif avg >= 40:
        return "C"
    else:
        return "Fail"


def add_student():
    try:
        roll = int(input("Enter Roll Number: "))
        name = input("Enter Name: ")

        marks = {}

        marks["Math"] = int(input("Enter Math Marks: "))
        marks["Science"] = int(input("Enter Science Marks: "))
        marks["English"] = int(input("Enter English Marks: "))

        avg = sum(marks.values()) / len(marks)
        grade = calculate_grade(avg)

        student = (roll, name, marks, grade)
        students.append(student)

        with open("result.txt", "a") as file:
            file.write(f"{roll},{name},{avg},{grade}\n")

        print("Result Saved")

    except ValueError:
        print("Invalid input")


def show_results():
    for s in students:
        print("\nRoll:", s[0])
        print("Name:", s[1])
        print("Marks:", s[2])
        print("Grade:", s[3])


while True:
    print("\n1.Add Student")
    print("2.Show Result")
    print("3.Exit")

    ch = input("Enter choice: ")

    if ch == "1":
        add_student()

    elif ch == "2":
        show_results()

    elif ch == "3":
        break

    else:
        print("Invalid Choice")
```
